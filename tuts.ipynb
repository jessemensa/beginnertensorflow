```python
# Interactive sessions 
t1 = tf.constant(1.2) 
t2 = tf.constant(3.5) 
prod = tf.multiply(t1, t2) 
with tf.Session() as sess:
    print("Product: ", sess.run(prod)) 
```

```python
# Same result as above 
t1 = tf.constant(1.2) 
t2 = tf.constant(3.5) 
prod = tf.multiply(t1, t2) 
sess = tf.InteractiveSession() 
print('Product: ', prod.eval()) 
```
